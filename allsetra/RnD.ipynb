```python
import pandas as pd

```

```python
df = pd.read_excel("PositionsUnit-LongLat.xlsx", index_col = [0])
```

```python
#Convert Longs and Lats to floats. 

def cnvrtCoord(val: str) -> float:
    sgn = +1
    if val[0] == 'S' or val[0] == 'W':
        sgn = -1
    return float(val[1:]) * sgn  

df['long'] = [cnvrtCoord(x) for x in df['long'].tolist()]
df['lat'] = [cnvrtCoord(x) for x in df['lat'].tolist()]  
```

```python
df['DateTimeOfPosition'] = pd.to_datetime(df['DateTimeOfPosition'])
```

```python
import pandas as pd

# Assuming your DataFrame is called 'df' with columns 'Timestamp', 'Longitude', 'Latitude', and 'Ignition On/Off'

# Sort the DataFrame by Timestamp in ascending order
df.sort_values('DateTimeOfPosition', inplace=True)

# Initialize variables to track trip start and end points
trip_start = None
trip_end = None

# Initialize a list to store the trips
trips = []

# Iterate through the DataFrame rows
for index, row in df.iterrows():
    # Check if ignition is turned on
    if row['IgnitionOn']:
        # If trip has not started yet, mark current row as trip start
        if trip_start is None:
            trip_start = row
    else:
        # If trip has started and ignition is turned off, mark current row as trip end
        if trip_start is not None:
            trip_end = row
            # Append the trip start and end points to the list of trips
            trips.append((trip_start['DateTimeOfPosition'], trip_start['lat'], trip_start['long'], trip_end['DateTimeOfPosition'], trip_end['lat'], trip_end['long']))
            # Reset trip start and end points for the next trip
            trip_start = None
            trip_end = None

# Create a new DataFrame to store the trips
trips_df = pd.DataFrame(trips, columns=['Start Timestamp', 'Start Longitude', 'Start Latitude', 'End Timestamp', 'End Longitude', 'End Latitude'])

# Print the trips DataFrame
print(trips_df)
```

```python
import pandas as pd
from geopy.distance import distance

# Assuming your DataFrame is called 'df' with columns 'Timestamp', 'Longitude', and 'Latitude'

# Sort the DataFrame by Timestamp in ascending order
df.sort_values('DateTimeOfPosition', inplace=True)

# Initialize variables to track trip start and end points
trip_start = None
trip_end = None

# Initialize a list to store the trips
trips = []

# Iterate through the DataFrame rows
for index, row in df.iterrows():
    # Check if trip has not started yet
    if trip_start is None:
        trip_start = row
    else:
        # Calculate distance between current point and the previous point
        dist = distance((trip_start['long'], trip_start['lat']), (row['long'], row['lat'])).meters
        # Calculate duration between current point and the previous point
        duration = (row['DateTimeOfPosition'] - trip_start['DateTimeOfPosition']).total_seconds()

        # If the distance is above a certain threshold, consider it as the car moving
        if dist > 50 and duration > 600:  # Adjust the distance threshold as per your requirements
            # Update trip end point
            trip_end = row
        else:
            # If the distance is below the threshold, consider it as the car stopped
            # Append the trip start and end points to the list of trips
            if trip_end is not None:
                trips.append((trip_start['DateTimeOfPosition'], trip_start['lat'], trip_start['long'], trip_end['DateTimeOfPosition'], trip_end['lat'], trip_end['long']))
            # Reset trip start and end points for the next trip
            trip_start = row
            trip_end = None

# If the last trip did not end, consider the last row as the trip end
if trip_start is not None and trip_end is None:
    trip_end = df.iloc[-1]
    trips.append((trip_start['DateTimeOfPosition'], trip_start['lat'], trip_start['long'], trip_end['DateTimeOfPosition'], trip_end['lat'], trip_end['long']))

# Create a new DataFrame to store the trips
trips_df = pd.DataFrame(trips, columns=['Start Timestamp', 'Start Longitude', 'Start Latitude', 'End Timestamp', 'End Longitude', 'End Latitude'])

# Print the trips DataFrame
print(trips_df)
```
